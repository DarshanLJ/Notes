```python
import requests,json
apikey="9c68bade3cc144767468cc951c78d0fa"
city=input("enter name ")
url="http://api.openweathermap.org/geo/1.0/direct?q="+city+"&appid="+apikey
response=requests.get(url)
coordinates=response.json()
print(coordinates)
```

```python
lat=coordinates[0]["lat"]
print(lat)
```

```python
lon=coordinates[0]["lon"]
print(lon)
```

```python
# current weather data
url1="https://api.openweathermap.org/data/2.5/weather?lat="+str(lat)+"&lon="+str(lon)+"&units=metric&appid="+apikey
response=requests.get(url1)
current_weather=response.json()
print(json.dumps(current_weather,indent=1))
```

```python
# qair pollution api
url2="http://api.openweathermap.org/data/2.5/air_pollution?lat="+str(lat)+"&lon="+str(lon)+"&appid="+apikey
response=requests.get(url2)
current_pollution=response.json()
print(json.dumps(current_pollution,indent=1))
```

```python
aqi=current_pollution["list"][0]["main"]["aqi"]
aqi
if aqi==1:
    print("Good")
elif aqi==2:
    print("Fair")
elif aqi==3:
    print("moderate")
elif aqi==4:
    print("poor")
elif aqi==5:
    print("verypoor")
```

```python
# 5 day 3 hour forcast
url3="http://api.openweathermap.org/data/2.5/forecast?lat="+str(lat)+"&lon="+str(lon)+"&appid="+apikey
response=requests.get(url3)
forcast=response.json()
print(json.dumps(forcast,indent=1))
```

```python
D = {"date_time" : [] ,"temp" : [] ,"pressure" : [] ,"humidity" : [],"description" : []}
for i in forcast ["list"]:
    D["date_time"].append(i["dt_txt"])
    D["temp"].append(i["main"]["temp"])    
    D["pressure"].append(i["main"]["pressure"])
    D["humidity"].append(i["main"]["humidity"])    
    D["description"].append(i["weather"][0]["description"])    
```

```python
import pandas as pd
df=pd.DataFrame(D)
df
```
